1. Import the necessary packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

2. Load the salary.csv file

```python
df = pd.read_csv('salary.csv')
```

3. Print the first 5 rows of the dataset

```python
display(df.head())
```

4. Display the general information of the dataframe

```python
df.info()
```

5. Visualize the data by creating a scatter plot

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
plt.show()
```

6. Split the dataset into a training set and test set

```python
X = df[['YearsExperience']]
y = df['Salary']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

7. Create an new instance of a linear regression model

```python
model = LinearRegression()
```

8. Train and test the model using the training set and test set

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

9. Print the equation of the line

- Display the slope
- Display the intercept
- Display the equation of the line

```python
# Display the slope
slope = model.coef_[0]
print(f"Slope: {slope}")

# Display the intercept
intercept = model.intercept_
print(f"Intercept: {intercept}")

# Display the equation of the line
print(f"Equation of the line: Salary = {slope:.2f} * YearsExperience + {intercept:.2f}")
```

10. Draw the Regression Line

```python
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.scatter(X_test, y_test, color='green', label='Test data')
plt.plot(X_train, model.predict(X_train), color='red', label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary with Regression Line')
plt.legend()
plt.show()
```

11. Compute the Sum of Square Error (SSE)

```python
# Compute the Sum of Square Error (SSE)
sse = mean_squared_error(y_test, y_pred) * len(y_test)
print(f"Sum of Square Error (SSE): {sse}")
```

12. Compute the $R^2$

```python
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R^2): {r2}")
```
